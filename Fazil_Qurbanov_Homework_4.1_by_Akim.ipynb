```python

'''Question 3
Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price.
'''
class ShoppingCart():
    def __init__(self):
        self.products = []

    def add_product(self, product, quantity, price):
        item = (product, quantity, price)
        self.products.append(item)
        return item
    
    def remove_item(self, product_name):
        for item in self.products:
            if item[0]==product_name:
                self.products.remove(item)
                

    def calculate_total_price(self):
        total = 0
        for i in self.products:
            total += (i[1]*i[2])
            
        return total



pr = ShoppingCart()
pr.add_product("Table", 10, 20)
pr.add_product("Phone", 20, 10)
pr.add_product("Bag", 30, 10)

pr.remove_item("Bag")
print("ShoppingCart updated:")
for item in pr.products:
    print(f"{item[0]} - {item[1]}, {item[2]} AZN")

print("ShoppingCart total value: ")
total_value_stock = pr.calculate_total_price()
print(total_value_stock)
    

    
```
